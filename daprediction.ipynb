```python
import numpy as np
import pandas as pd
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
data1 = pd.read_csv("/Users/rjy/Desktop/20240516 XIY西安落地/data1.csv")
data1
```

```python
data2 = pd.read_csv("/Users/rjy/Desktop/20240516 XIY西安落地/data2.csv")
data2
```

```python
#进行合并
merged_data = pd.merge(data1, data2, on='flt_id', how='inner')
merged_data
```

```python
print(merged_data.columns)
```

```python

merged_data_dropped = merged_data.drop(columns=['flt_id','month_of_year', 'week_of_year', 'day_of_week', 'day_of_year', 'minutes_of_day'])

merged_data['实际降落'] = pd.to_datetime(merged_data['实际降落'])
merged_data['北京时间'] = pd.to_datetime(merged_data['北京时间'])

merged_data['时间差分钟'] = (merged_data['实际降落'] - merged_data['北京时间']).dt.total_seconds() / 60

target_on_block = merged_data['时间差分钟']

features_on_block = merged_data_dropped.select_dtypes(include=[np.number])


X_train, X_test, y_train, y_test = train_test_split(features_on_block, target_on_block, test_size=0.3, random_state=42)


model = RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(X_train, y_train)
model
y_pred = model.predict(X_test)
mse = mean_squared_error(y_test, y_pred)

feature_importance = model.feature_importances_
feature_importance_dict = {feature: importance for feature, importance in zip(features_on_block.columns, feature_importance)}


mse, feature_importance_dict

```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

#预测时间差是否小于30分钟并计算准确率
merged_data2['flt_ata_dttm_bj'] = pd.to_datetime(merged_data2['flt_ata_dttm_bj'])
merged_data2['source_time'] = pd.to_datetime(merged_data2['source_time'])

merged_data2['时间差分钟'] = (merged_data2['flt_ata_dttm_bj'] - merged_data2['source_time']).dt.total_seconds() / 60

merged_data2['时间差小于30'] = (merged_data2['时间差分钟'] < 30).astype(int)

features = merged_data2.drop(columns=['flt_id', 'month_of_year', 'week_of_year', 'day_of_week', 'day_of_year', 'flt_ata_dttm_bj', 'source_time', '时间差分钟', '时间差小于30'])
features = features.select_dtypes(include=[np.number])

target = merged_data2['时间差小于30']

X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.3, random_state=42)

model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
print("准确率:", accuracy)
```
